### Question 2: Subset the DataFrame (1 Marks)

### Question 5: Find Top 5 Affected Countries (2 Marks)

From the original `df_covid` DataFrame, find the 5 countries with the highest total `Cumulative_cases`. Group the data by `Country`, find the maximum `Cumulative_cases` for each, sort the results, and get the top 5. Print the resulting Series.

**Expected Output:** A Pandas Series showing the top 5 countries as the index and their max cumulative cases as the values.

### Question 6: Calculate Daily Global Cases (2 Marks)

Using the `df_covid_trimmed` DataFrame, calculate the total number of `New_cases` for each date across all countries. Store this in a new DataFrame called `daily_global_cases` and print its head.

### Question 7: Aggregate Data by Quarter (2 Marks)

Aggregate the `New_cases` and `New_deaths` from `df_covid_trimmed` on a quarterly basis. First, create a 'Quarter' column, then group by this column and sum the cases and deaths. Display the head of the resulting DataFrame.

### Question 10: Pivot Table for Monthly Cases by Region (2 Marks)

Create a pivot table to show the total `New_cases` for each `WHO_region` aggregated by month. First, create a 'Month_Year' column. Then, create the pivot table with `WHO_region` as the index, `Month_Year` as columns, and the sum of `New_cases` as values. Display the resulting pivot table.

**Expected Output:** A table with regions as rows, month-years as columns, and aggregated new cases as the cell values.

```python
# Write your answer here
df_help2 = df_covid
df_help2['Date_reported'] = pd.to_datetime(df_help2['Date_reported'])

df_help2['Month_Year'] = df_help2['Date_reported'].dt.to_period('M')

pivot = pd.pivot_table(
    df_help2,
    values='New_cases',
    index='WHO_region',
    columns='Month_Year',
    aggfunc='sum'
)

print(pivot.head())
```

### Question 12: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

### Question 13: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
# Write your answer here
from sklearn.metrics import f1_score
import numpy as np

# Step 1: Load dataset
digits = load_digits()

X = digits.data
y_true = digits.target

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=42)
y_cluster = kmeans.fit_predict(X)

# Step 3: Map cluster labels to actual labels (majority voting)
mapped_labels = np.zeros_like(y_cluster)

for cluster in range(10):
    mask = (y_cluster == cluster)
    true_labels = y_true[mask]

    if len(true_labels) == 0:
        continue

    # majority label using numpy
    most_common = np.bincount(true_labels).argmax()
    mapped_labels[mask] = most_common

# Step 4: Evaluate using F1 score
f1 = f1_score(y_true, mapped_labels, average='macro')

print("F1 Score:", f1)








```
